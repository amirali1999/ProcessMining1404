```python
from apps.preprocesing.pipline.log_preprocess import LogPreprocessor
pre = LogPreprocessor()
pre.load("Sepsis Cases - Event Log.xes.gz")
```

```python
#pre.debug = True
pre.smart_clean(scope="all",inplace=True)
```

```python
pre.anonymize_column("case:concept:name", mode="numeric", inplace=True)
pre.drop_columns(["concept:name","time:timestamp","case:concept:name"],keep_mode=True)
```

```python
pre.drop_constant_columns(null_threshold=0.9)
```

```python
pre.drop_columns()
```

```python
pre.sort_by_columns(["case:concept:name","time:timestamp"],ascending=True,inplace=True)
```

```python
pre.anonymize_column("case:concept:name",mode="numeric",inplace=False)
```

```python
pre.anonymize_column("time:timestamp",mode="alphabetic",inplace=False)
```

```python
pre.rename_column("case:concept:name","ID",inplace=False)
```

```python
pre.df
```

```python
pre.merge_columns(columns=["concept:name","lifecycle:transition"],new_column="my:column",delimiter="_",inplace=False)
```

```python
pre.convert_column_type("time:timestamp","datetime",on_fail="delete",datetime_formats = ["%Y-%m-%d", "%d/%m/%Y", "%Y-%m-%d %H:%M:%S"], inplace=False)
```

```python
pre.convert_column_type("case:concept:name","numeric",on_fail="default",default_value="******",inplace=False)
```

```python
pre.df
```

```python
pre.remove_duplicates()
```

```python
pre.remove_duplicates(["case:concept:name","concept:name"],keep="first",inplace=False)
```

```python
pre.remove_by_condition("`case:concept:name` != 'A' and `case:concept:name` != 'B' ",inplace=False)
```

```python
pre.df
```

```python
pre.replace_values("concept:name",["ER Registration","Leucocytes"],new_value="******",inplace=False)
```

```python
pre.replace_values(columns="concept:name",to_replace=["JOIN-PAT","SET STATUS","CODE ERROR","CHANGE END","ZDBC_BEHAN"],new_value="Unknown",inplace=True)
```

```python
pre.handle_null_values("case:concept:name",on_null="delete",inplace=False)
```

```python
pre.filter_by_range_and_allowed("time:timestamp",range_min="2014-10-22 00:00:00+00:00",range_max="2014-10-23 00:00:00+00:00",on_outside="delete",inplace=False)
```

```python
pre.handle_outliers("time:timestamp",deviation_percent=50,groupby_column="case:concept:name",on_outside="delete",inplace=False)
```

```python
pre.unique_summary("concept:name")
```

```python
pre.reset_to_original()
```

```python
df=pre.suggest_corrections("concept:name",["Leucocytes","CRP","ER Registration","LacticAcid","Admission NC","ER Triage"],threshold=15)
```

```python
pre.apply_corrections("concept:name",df,inplace=False)
```
